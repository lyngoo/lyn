```python
iris=sns.load_dataset("iris")
iris.head()
```

```python
vals=iris['sepal_length'].values
sns.kdeplot(vals)
plt.title("kdeplot for sepal_length")
plt.show()
```

```python
vals=iris['petal_length'].values
sns.kdeplot(vals)
plt.title("kdeplot for petal_length")
plt.show()
```

```python
vals=iris['petal_length'].values
sns.distplot(vals,kde=True,rug=True)
plt.title("displot for petal_length")
plt.show()
```

```python
sns.displot(vals,kde=True,rug=True)
plt.title("displot for petal_length")
plt.show()
```

```python
sns.histplot(vals,kde=True)
plt.title("histplot for petal_length")
plt.show()
```
